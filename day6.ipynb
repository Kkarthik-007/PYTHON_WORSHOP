```python
#generate prime numbers from 2 to n    
#input-->n=10
#output-->2 3 5 7
#1.create a boolean list to the size of n+1
#2.mark all the index as true
#3.mark the multiples of 2 and 3 as false
#4.start from i=2 to i=n+1 print all the true index
##SIEVE OF ERATOSTHENES
n=50
prime=[True]*(n+1)
i=2
while i*i<=n:
    if prime[i]:
        for j in range(i*i,n+1,i):
            prime[j]=False
        i=i+1
    for i in range(2,n+1):
        if prime[i]:
         print(i,end=" ")        
```
