```python
n=3
fact=1
for i in range(1,n+1):
    fact=fact*i
    print(fact)
```

```python
#reverse an integer
#input=234
#output=432
num=234
rev=0
while num>0:
    rem=num%10
    rev=(rev*10)+rem
    num=num//10
print(rev)    
```

```python
#pattern matching
#n=4
#*
#**
#***
#****
#n represents the number of rows and rows represents the number of stars
n=4
for i in range(1,n+1):
    for j in range(i):
        print("*",end="")
    print()    
```

```python
n=4
for i in range(n,0,-1):
    for j in range(i):
        print("*",end="")
    print()
```
