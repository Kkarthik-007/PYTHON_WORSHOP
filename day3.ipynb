LOGICAL OPERATORS

```python
#Print greatest of three numbers
a= float(input("Enter a value:"))
b= float(input("Enter b value:"))
c= float(input("Enter c value:"))
if (a>b and a>c):
    print("A is greater")
elif (b>a and b>c):
    print("B is greater")
else:
    print("C is greater")    
```

```python
#Print greatest of four numbers
a= float(input("Enter a value:"))
b= float(input("Enter b value:"))
c= float(input("Enter c value:"))
d= float(input("Enter d value:"))
if (a>b and a>c and a>d):
    print("A is greater")
elif (b>a and b>c and b>d):
    print("B is greater")
elif (c>a and c>b and c>d):
    print("C is greater")    
else:
    print("D is greater")    
```

```python
#CHECK WHETHER A GIVEN CHARACTER IS VOWEL OR CONSONANT(LOWERCASE)
vowel = ("a","e","i","o","u")
ch=(input("Enter the character:")).lower()
if(ch in vowel):
    print("IT is a vowel")
else:
    print("it is a consonant")    
```

```python
#NESTED IF-ELSE
username = ("Admin")
password = ("1234")
user=(input("Enter the username:"))
passw=(input("Enter the password:"))
if(user in username):
    if(passw in password):
        print("login Successful")
    else:
        print("invalid password")
else:
    print("invalid username")           
```

```python
citizen = ("Indian")

name=(input("Enter the citizen:"))
age=int(input("Enter the Age:"))
if(name in citizen):
    if(age > 18):
        print("eligible for vote")
    else:
        print("not eligible")
else:
    print("invalid citizenship")           
```

```python
#for loop in strings
str="python"
for ch in str:
    print(ch)
```

```python
#nested for loop(for loop inside another for loop)
for i in range(0,3):
    for j in range(0,3):
        print(j,end=",")
      
```

```python
#while loop number of iteration is not known
#1.initialization
#2.condition
#3.statements
#4.increment or decrement
#repeat until condition fails
i=0
while i<5:
    print(i)
    i=i+1
         
```

```python
#hailstone sequence
n=int(input("enter the n value:"))
while n!=1:
    print(n,end=",")
    if(n%2==0):
        n=n//2
    else:
        n=(3*n)+1
print(1)            
```
