```python
#CHECK WHETHER A NUMBER IS POSITIVE OR NEGATIVE
num = float(input("Enter a number:"))
if (num>0):
 print("positive num")
elif(num==0):
 print("Num is Zero") 
else:
 print("negative num") 
```

```python
#CHECK WHETHER A NUM IS EVEN OR ODD
num = float(input("Enter a number:"))
if(num%2==0):
 print("Even")
else:
 print("Odd")
  
```

```python
#Print greatest of three numbers
a= float(input("Enter a value:"))
b= float(input("Enter b value:"))
c= float(input("Enter c value:"))
if (a>b and a>c):
    print("A is greater")
elif (b>a and b>c):
    print("B is greater")
else:
    print("C is greater")    
```

```python
#Print greatest of four numbers
a= float(input("Enter a value:"))
b= float(input("Enter b value:"))
c= float(input("Enter c value:"))
d= float(input("Enter d value:"))
if (a>b and a>c and a>d):
    print("A is greater")
elif (b>a and b>c and b>d):
    print("B is greater")
elif (c>a and c>b and c>d):
    print("C is greater")    
else:
    print("D is greater")    
```

```python
#CHECK WHETHER A GIVEN CHARACTER IS VOWEL OR CONSONANT(LOWERCASE)
vowel = ("a","e","i","o","u")
ch=(input("Enter the character:")).lower()
if(ch in vowel):
    print("IT is a vowel")
else:
    print("it is a consonant")    
```

```python
#NESTED IF-ELSE
username = ("Admin")
password = ("1234")
user=(input("Enter the username:"))
passw=(input("Enter the password:"))
if(user in username):
    if(passw in password):
        print("login Successful")
    else:
        print("invalid password")
else:
    print("invalid username")           
```
