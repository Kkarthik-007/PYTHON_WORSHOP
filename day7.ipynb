```python
class Animal:
    def eat(self):
        print("animal eats")
a=Animal()
a.eat()        
```

```python
#inheritannce-->inheriting one class to another class
class Animal:#parent class
    def eat(self):
        print("animal eats")
class lion(Animal):#child class
    def king(self):
        print("king of the jungle")
l=lion()
l.eat()
l.king()                
```

```python
#Single inheritance
#Multi level inheritance
#Hierarchical inheritance
#Multiple inheritance
#Hybrid inheritance
```

```python
#multi level inheritance
#grand parent-->parent-->child
class Animal:
    def eat(self):
        print("animal eats")
class Lion(Animal):
    def king(self):
        print("king of the jungle")
class Cub(Lion):
    def sleep(self):
        print("cubs sleep")
c=Cub()
c.sleep()
c.king()
c.eat()        
```

```python
#Hierrarchical inheritance
class Animal:
    def eat(self):
        print("animal eats")
class Lion(Animal):
    def king(self):
        print("king of the jungle")
class Dog(Animal):
    def guard(self):
        print("guards the home")
d=Dog()
d.eat()
d.guard()
l=lion()
l.eat()
l.king()
```

```python
#Multiple inheritance
class Animal:
    def eat(self):
        print("animal eats")
class WildAnimal:
    def hunt(self):
        print("Animal hunts")
class Lion(Animal,WildAnimal):
    def King(self):
        print("King of the jungle")

l=Lion()
l.King()
l.hunt()
l.eat()
```

ABSTRACTION

```python
from abc import ABC,abstractmethod
class Vehicle(ABC):
    @abstractmethod#declared but not defined
    def start(self):
        pass
class Car(Vehicle):
    def start(self):
        print("car is starting")
c=Car() 
c.start()         
```
