BREAKING STATEMENTS(break,continue,pass)

```python
#break-->used to break a loop
for i in range(0,5):
    if(i==3):
        break
    print(i)
```

```python
#continue-->used to skip the iteration
for i in range(0,5):
    if(i==2 or i==3):
        continue
    print(i)    
```

```python
#pass-->do nothing(palce holder)
for i in range(0,5):
    pass
```

```python
#find the first non repeating character in a string
#input-->"swiss"
#output-->"w"
str="swiss"
for ch in str:
    if str.count(ch)==1:
        print(ch)
        break
    else:
        print("no non repeated character")
```
